## Create Delta Lake Tables

```sql
USE CATALOG databricksdemo
```

```sql
CREATE TABLE employees
  (id INT, name STRING, salary DOUBLE);
```


## Catalog Explorer

Check the created **employees** table in the **Catalog** explorer.
