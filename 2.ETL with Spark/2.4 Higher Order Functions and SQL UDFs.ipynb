## Filtering Arrays

```sql
select order_id, books, filter(books, i -> i.quantity >= 2) as multiple_copies
from orders limit 3
```

```sql
select order_id, multiple_copies from
(select order_id, books, filter(books, i -> i.quantity >= 2) as multiple_copies
from orders)
where size(multiple_copies) > 0 limit 3;
```

## Transforming Arrays

```sql
select 
  order_id, 
  books,
  transform(
    books, 
    i -> cast(i.subtotal * 0.8 as int)
  ) as subtotal_after_discount
from orders limit 3
```

## User Defined Functions (UDF)

```sql
create or replace function get_url(email string)
returns string
return concat("https://www.", split(email, '@')[0])
```

```sql
select * from customers limit 3
```

```sql
select email, get_url(email) from customers limit 3
```

```sql
describe function get_url
```

```sql
describe function extended get_url
```

```sql
Create or replace function site_type(email string)
returns string
return case 
          when email like "%.com" then "Commercial busiess"
          when email like "%.org" then "Non-profits organization"
          when email like "%.edu" then "Educational institution"
          else concat("Unknow extenstion for domain: ", split(email, "@")[1])
       end;
```

```sql
select email, get_url(email), site_type(email) from customers limit 3
```

```sql
DROP FUNCTION get_url;
DROP FUNCTION site_type;
```
